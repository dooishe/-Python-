Задание 1

```python
sentence = input("Введите предложение: ")
words = sentence.lower().split()

cleaned_words = []
for word in words:
    cleaned_word = ''.join([char for char in word if char.isalpha()])
    if cleaned_word:
        cleaned_words.append(cleaned_word)

unique_words = set(cleaned_words)

print("Уникальные слова в предложении:")
for word in sorted(unique_words):
    print(word)
```

```python
s = input("Введите строку: ")

max_count = 0
current_count = 0

for char in s:
    if char == 'а':
        current_count += 1
        if current_count > max_count:
            max_count = current_count
    else:
        current_count = 0

print(f"Самая длинная последовательность букв 'а': {max_count}")
```

```python
from collections import defaultdict

text = input("Введите строку: ")
count = defaultdict(int)

# Подсчет символов
for char in text:
    count[char] += 1

# Фильтрация и сохранение порядка
result = [char for char in text if count[char] == 1]
print(''.join(result))
```

```python
import struct
import random
import math

def write_numbers_to_file(filename, N):
    with open(filename, 'wb') as f:
        for _ in range(N):
            num = random.uniform(-100, 100)
            f.write(struct.pack('d', num))

def find_max_abs_odd(filename):
    max_abs = 0.0
    with open(filename, 'rb') as f:
        index = 1
        while True:
            bytes_data = f.read(8)
            if not bytes_data:
                break
            num = struct.unpack('d', bytes_data)[0]
            if index % 2 != 0:
                current_abs = abs(num)
                if current_abs > max_abs:
                    max_abs = current_abs
            index += 1
    return max_abs

filename = 'numbers.bin'
N = int(input("Введите количество чисел N: "))

write_numbers_to_file(filename, N)

result = find_max_abs_odd(filename)
print(f"Наибольшее из значений модулей компонент с нечетными номерами: {result}")

with open(filename, 'rb') as f:
    print("\nВсе числа в файле:")
    while True:
        bytes_data = f.read(8)
        if not bytes_data:
            break
        num = struct.unpack('d', bytes_data)[0]
        print(num, end=' ')
```

```python
def find_most_expensive_constructor(filename):
    max_price = 0
    constructor_name = ""

    try:
        with open(filename, 'r', encoding='utf-8') as file:
            for line in file:
                parts = line.strip().split(';')
                if len(parts) < 3:
                    continue

                name = parts[0].strip()
                try:
                    price = float(parts[1].strip())
                except ValueError:
                    continue

                if "конструктор" in name.lower():
                    if price > max_price:
                        max_price = price
                        constructor_name = name

    except FileNotFoundError:
        print(f"Файл {filename} не найден")
        return None

    if max_price > 0:
        return {"name": constructor_name, "price": max_price}
    else:
        return None

filename = "Assort.txt"
result = find_most_expensive_constructor(filename)

if result:
    print(f"Самый дорогой конструктор: '{result['name']}', стоимость: {result['price']} руб.")
else:
    print("Конструкторы не найдены в файле или файл пуст.")
```

```python
import re

def find_and_calculate_expressions(input_file):

    try:
        with open(input_file, 'r', encoding='utf-8') as f:
            text = f.read()

        pattern = r'(-?\d+\.?\d*)\s*([+\-*/])\s*(-?\d+\.?\d*)'
        expressions = re.findall(pattern, text)

        if not expressions:
            print("Арифметические выражения не найдены.")
            return

        print("Найденные арифметические выражения и их результаты:")
        for expr in expressions:
            num1, op, num2 = expr
            num1 = float(num1)
            num2 = float(num2)

            try:
                if op == '+':
                    result = num1 + num2
                elif op == '-':
                    result = num1 - num2
                elif op == '*':
                    result = num1 * num2
                elif op == '/':
                    result = num1 / num2

                print(f"{num1}{op}{num2} = {result}")
            except ZeroDivisionError:
                print(f"{num1}{op}{num2} = Ошибка: деление на ноль")

    except FileNotFoundError:
        print(f"Ошибка: файл {input_file} не найден.")
    except Exception as e:
        print(f"Произошла ошибка: {str(e)}")

if __name__ == "__main__":
    input_filename = "input.txt"

    with open(input_filename, 'r', encoding='utf-8'):
            pass

    print(f"Создан тестовый файл {input_filename} с примером данных")

    find_and_calculate_expressions(input_filename)
```
